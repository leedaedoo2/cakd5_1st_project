```python
prod_df = pd.merge(df2,df1,on='소분류코드')

prod_matrix = prod_df.pivot_table('구매횟수',index = '고객번호', columns = '소분류명',aggfunc='sum')

prod_matrix = prod_matrix.fillna(0)
prod_matrix
```

```python
prod_matrix_T = prod_matrix.transpose()
prod_matrix_T
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(prod_matrix_T,prod_matrix_T)

item_sim_df = pd.DataFrame(data = item_sim, index = prod_matrix.columns, columns = prod_matrix.columns)

print(item_sim_df.shape)
item_sim_df
```

```python
def predict_prod_topsim(prod_arr,item_sim_arr,n=20):
    pred = np.zeros(prod_arr.shape)
    
    for col in range(prod_arr.shape[1]):
        
        top_n_times = [np.argsort(item_sim_arr[:,col])[:-n-1:-1]]
        
        for row in range(prod_arr.shape[0]):
            pred[row,col] = item_sim_arr[col,:][top_n_times].dot(prod_arr[row,:][top_n_times].T)
            pred[row,col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_times]))
            
    return pred

prod_pred = predict_prod_topsim(prod_matrix.values,item_sim_df.values,n=20)

prod_pred_matrix = pd.DataFrame(data = prod_pred, index = prod_matrix.index, columns = prod_matrix.columns)

prod_pred_matrix
```

```python
def get_zero_prod(prod_matrix,userId):
    user_prod = prod_matrix.loc[userId,:]
    
    already_buy = user_prod[user_prod > 0].index.tolist()
    
    prod_list = prod_matrix.columns.tolist()
    
    unbuy_list = [prod for prod in prod_list if prod not in already_buy]
    
    return unbuy_list

def recomm_prod_by_userid(pred_df, userId, unbuy_list, top_n=10):
    
    recomm_prod = pred_df.loc[userId,unbuy_list].sort_values(ascending=False)[:top_n]
    
    return recomm_prod
```
