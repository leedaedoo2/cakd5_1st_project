```python
col_list = list(df.columns[:-5:-1])
col_list
```

```python
for col in col_list:
    q1 = np.percentile(df[col],25)
    q2 = np.percentile(df[col],50)
    q3 = np.percentile(df[col],75)
    df.loc[(df[col]<q1),col] = 1
    df.loc[(df[col]>=q1)&(df[col]<q2),col] = 2
    df.loc[(df[col]>=q2)&(df[col]<q3),col] = 3
    df.loc[(df[col]>=q3),col] = 4
    
    
df
```

### RFM 에서 M 금액의 부분이 상간관계가 눈에 띄게 높음
### 가공,신선 식품과 빈도의 상관관계 높음

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:-1]
y = df['15년 하반기 구매총액']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
dt_pred = dt.predict(X_test)
print('결정트리 정확도:',accuracy_score(y_test,dt_pred))
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
dt.feature_importances_
dt_series = pd.Series(dt.feature_importances_,index=X_train.columns)
dt_series.sort_values(ascending=False)
dt_series.sort_values().plot(kind='barh')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
lr_clf = LogisticRegression()
lr_clf.fit(X_train,y_train)
lr_preds = lr_clf.predict(X_test)
mse =mean_squared_error(y_test,lr_preds)
print(accuracy_score(y_test,lr_preds))
print(mse)
```

```python
from lightgbm import LGBMClassifier
lgbm = LGBMClassifier(n_estimators=300)
evals =[(X_test,y_test)]
lgbm.fit(X_train,y_train,early_stopping_rounds=100,eval_metric ='logloss',eval_set = evals,verbose=True)
preds = lgbm.predict(X_test)
pred_proba = lgbm.predict_proba(X_test)[:,1]
print('lgbgm 의 정확도',accuracy_score(y_test,preds))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

rf = RandomForestRegressor(random_state=0)
rf.fit(X_train,y_train)
neg_mse_score = cross_val_score(rf,X,y,scoring = 'neg_mean_squared_error',cv = 5)

rmse = np.sqrt(-1*neg_mse_score)
avg_rmse = np.mean(rmse)
print(avg_rmse)
```
