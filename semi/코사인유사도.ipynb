```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))

doc1 = np.array([0,1,1,1])
doc2 = np.array([1,0,1,1])
doc3 = np.array([2,0,2,2])

print('문서 1과 문서2의 유사도 :',cos_sim(doc1, doc2))
print('문서 1과 문서3의 유사도 :',cos_sim(doc1, doc3))
print('문서 2와 문서3의 유사도 :',cos_sim(doc2, doc3))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_json('./melon_dataset/test.json')
data.head()
```

```python
import numpy as np
tag_names = data.tags
tag_name2 = list(np.full(len(tag_names),''))

#print(tag_name2)
for i in range(len(tag_names)):
#     print(tag_names[i])
    tmp = ''
    for j in range(len(tag_names[i])):
        tmp = tmp + tag_names[i][j]
#         print(tmp)
        if j != len(tag_names[i]) - 1:
            tmp = tmp + ', '
#         print(tmp)
#     print(tmp)
    tag_name2[i] = tmp
    print(tag_name2[i])
tag_name2
```

```python
data.tags = tag_name2
data
```

```python
tag_lists = list(data['tags'])
tag_lists.append('겨울')
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(tag_lists).toarray()
print('TF-IDF 행렬의 크기(shape) :',tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print('코사인 유사도 연산 결과 :',cosine_sim.shape)
```
