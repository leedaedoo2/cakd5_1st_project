태그의 종류 추출 > 각 태그별 컬럼을 만들어 True값 반환 > 새 df 생성

```python
# 태그 종류 추출
tag_list = []
for i in range(len(tag_arr)):
    if tag_arr[i] == []:
        continue
    else:
        for j in range(len(tag_arr[i])):
            if tag_arr[i][j] not in tag_list:
                tag_list.append(tag_arr[i][j])
            else:
                continue
                
print(tag_list)
print(len(tag_list))
```

```python
import re

p = re.compile('\S{0,10}\S{0,10}')

for i in range(len(tag_list)):
    m = re.findall(p,tag_list[i])
    if m == []:
        continue
    else:
        print(m)
```

```python
tag_cold = ['추위','추워','추움','추운날씨','추운','추운날','쌀쌀한','쌀쌀한날씨']
tag_cool = ['시원한','시원','시원해지는','시원하게','쿨여름']
tag_warm = ['따듯','따뜻하게','따뜻한','따듯함','따스한','따뜻','뜨뜻미지근한','포근한']
tag_hot = ['뜨거운','무더위','더위','더울때','더위사냥','더운날','더워']
tag_rain = ['비오는날','가을비','비','비올때','비가올때','봄비','겨울비']
tag_snow = ['첫눈','함박눈','흰눈','눈','눈오는날','눈내리는']
tag_sunny = ['맑은','맑은날','밝은']
tag_cloudy = ['먹구름','흐린날','흐림']
tag_cat = [tag_cold, tag_cool, tag_warm, tag_hot, tag_rain, tag_snow, tag_sunny, tag_cloudy]
print(tag_cat)
```

```python
# 태그 일치 여부를 저장
cold  = np.zeros(len(tag_arr))
cool = np.zeros(len(tag_arr))
warm = np.zeros(len(tag_arr))
hot = np.zeros(len(tag_arr))
rain = np.zeros(len(tag_arr))
snow = np.zeros(len(tag_arr))
sunny = np.zeros(len(tag_arr))
cloudy = np.zeros(len(tag_arr))
cat = np.array([cold,cool,warm,hot,rain,snow,sunny,cloudy])
```

```python
for i in range(len(tag_arr)):
    for j in range(len(tag_arr[i])):
        for k in range(len(tag_cat)):
            if tag_arr[i][j] in tag_cat[k]:
                if cat[k][i] ==1:
                    continue
                else:
                    cat[k][i] = 1
            else:
                continue
                
cat = cat.transpose()
df1 = pd.DataFrame(cat,columns = ['cold','cool','warm','hot','rain','snow','sunny','cloudy'])
df1
```
