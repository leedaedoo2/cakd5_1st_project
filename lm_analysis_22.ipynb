```python
import numpy as np
```

```python
import pandas as pd

lm_df = pd.read_csv('./data1.csv',index_col = 0)
lm_df.fillna(0,inplace=True)
lm_df
```

```python
lm_df.describe()
```

```python
col_list = list(lm_df.columns[10:])
col_list
```

```python
for col in col_list:
    col_arr = np.array(list(set(lm_df[col])))
    q1 = np.percentile(col_arr,25)
    q2 = np.percentile(col_arr,50)
    q3 = np.percentile(col_arr,75)
    q4 = np.max(col_arr)
    lm_df.loc[(lm_df[col]==0),col]=0
    lm_df.loc[(lm_df[col]>0)&(lm_df[col]<q1),col]=1
    lm_df.loc[(lm_df[col]>=q1)&(lm_df[col]<q2),col]=2
    lm_df.loc[(lm_df[col]>=q2)&(lm_df[col]<q3),col]=3
    lm_df.loc[(lm_df[col]>=q3)&(lm_df[col]<=q4),col]=4
    print(q1,q2,q3,q4)
```
