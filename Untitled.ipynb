태그의 종류 추출 > 각 태그별 컬럼을 만들어 True값 반환 > 새 df 생성

```python
# 태그 종류 추출
tag_list = []
for i in range(len(tag_arr)):
    if tag_arr[i] == []:
        continue
    else:
        for j in range(len(tag_arr[i])):
            if tag_arr[i][j] not in tag_list:
                tag_list.append(tag_arr[i][j])
            else:
                continue
                
print(tag_list)
print(len(tag_list))
```
